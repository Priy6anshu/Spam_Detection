```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/sms-spam-collection-dataset/spam.csv" , encoding='latin-1')
df.head()
```

```python
col = df.columns
```

```python
df.fillna(" " , inplace = True)
df["v2"] = df["v2"]+df["Unnamed: 2"]+df["Unnamed: 3"]+df['Unnamed: 4']
df["v2"]
```

```python
df = df.drop(["Unnamed: 2" , "Unnamed: 3" , 'Unnamed: 4'] , axis = 1)
```

```python
target = pd.get_dummies(df["v1"] , drop_first=True)
target.head()
```

```python
df = df.drop("v1" , axis =1)
df = pd.concat([df , target] , axis =1)
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.model_selection import train_test_split
X_train  , X_test , y_train, y_test = train_test_split(df.v2 , df.spam, test_size = 0.2)
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([('Vectorizer', CountVectorizer()), ('nb', MultinomialNB())])
pipe.fit(X_train , y_train)
```

```python
pipe.score(X_test , y_test)
```

```python
pipe.predict_proba(["The offer is valid only for today, so hurry up"])
```
